```python
#This program loops through all possible pairs of inputs and returns the length of the bracelet formed by each input.

import numpy as np
import matplotlib.pyplot as plt

n = 100

def find_sequence_length(start1, start2):
    seq = np.zeros(n, int)
    seq[0] = start1
    seq[1] = start2

    for i in range(n - 2):
        seq[i + 2] = seq[i + 1] + seq[i]
        seq[i + 2] = seq[i + 2] % 10

        if seq[i + 1] == start1 and seq[i + 2] == start2:
            return i + 1

    # If the loop completes without finding the condition, return -1
    return -1

# Initialize lists to store starting numbers and corresponding sequence lengths
start_list = list(range(10))  # All combinations for 0 to 9
lengths_list = {}

# Iterate through all combinations of starting numbers
for start1 in start_list:
    for start2 in start_list:
        length = find_sequence_length(start1, start2)
        
        # If the length is not in the dictionary, add it
        if (start1, start2) not in lengths_list:
            lengths_list[(start1, start2)] = [length]
        # If the length is in the dictionary, append it
        else:
            lengths_list[(start1, start2)].append(length)

# Print or visualize the possible lengths for each pair
for pair, lengths in lengths_list.items():
    print(f"Pair {pair}: Possible Lengths {lengths}")
```
