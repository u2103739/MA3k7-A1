```python
#Code proves that for every two starting possibilities the bracelet eventually returns to the two starting values to form a cycle.

import numpy as np
import matplotlib.pyplot as plt

n = 100

def find_sequence_length(start1, start2):
    seq = np.zeros(n, int)
    seq[0] = start1
    seq[1] = start2

    for i in range(n - 2):
        seq[i + 2] = seq[i + 1] + seq[i]
        seq[i + 2] = seq[i + 2] % 10

        if seq[i + 1] == start1 and seq[i + 2] == start2:
            return i + 1

    # If the loop completes without finding the condition, return -1
    return -1

# Initialize lists to store starting numbers and corresponding sequence lengths
start_numbers = list(range(10))  # Assuming you want to try all combinations for 0 to 9
sequence_lengths = []

# Iterate through all combinations of starting numbers
for start1 in start_numbers:
    for start2 in start_numbers:
        length = find_sequence_length(start1, start2)
        sequence_lengths.append(length)

# Reshape the lengths into a 2D array for plotting
lengths_array = np.array(sequence_lengths).reshape(len(start_numbers), len(start_numbers))

# Plot the lengths using a heatmap
plt.imshow(lengths_array, cmap='viridis', origin='lower', extent=[0, 9, 0, 9])
plt.colorbar(label='Sequence Length')
plt.xlabel('Starting Number 2')
plt.ylabel('Starting Number 1')
plt.title('Sequence Length for Different Starting Numbers')
plt.show()
```
